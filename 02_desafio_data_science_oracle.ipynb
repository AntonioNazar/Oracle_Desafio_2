# 0. Importando o dataset

```python
import pandas as pd
import plotly.express as px

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json"
```

```python
df = pd.read_json(url)
df.head()
```

# 1. Tratamento de dados inicial

```python
def normalizar_coluna(lista_coluna,df):
    for Coluna in lista_coluna:
        _ = pd.json_normalize(df[Coluna])
        df.drop(columns=Coluna, inplace=True)
        df = pd.concat([df, _], axis=1)
    return df

lista_coluna = ['customer', 'phone', 'internet', 'account']
df = normalizar_coluna(lista_coluna,df)
df.head()
```

```python
df.columns = df.columns.str.replace(r'(?<=[a-z0-9])([A-Z])|([.])', r'_\1', regex=True).str.lower()
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.loc[df['charges_total'] == ' ']
```

```python
# Dado que o ternure é 0, é perceptível que esses clientes são novos e ainda não possuem gastos totais, então vamos substituir os valores vazios por 0.0
df.loc[df['charges_total'] == ' ', 'charges_total'] = 0.0
df.loc[:,'charges_total'] = df.loc[:,'charges_total'].astype(float)
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.drop(columns=['customer_id'], inplace=True)
df
```

# 3. Relatório das informações

## Relatório de Análise de Evasão de Clientes (Churn)

### Introdução  

O presente relatório tem como objetivo analisar os fatores associados à evasão de clientes (churn) em uma base de dados de serviços de telecomunicações. A evasão de clientes representa um dos principais desafios estratégicos para empresas do setor, impactando diretamente a receita recorrente e os custos operacionais, uma vez que a aquisição de novos clientes tende a ser significativamente mais onerosa do que a retenção dos atuais.  

A análise busca identificar padrões comportamentais e características contratuais associadas a maiores taxas de churn, a fim de fornecer subsídios para ações estratégicas voltadas à retenção de clientes.

---

### Limpeza e Tratamento de Dados  

Inicialmente, os dados foram importados e submetidos a um processo de verificação de inconsistências, incluindo análise de valores ausentes, duplicados e possíveis erros de tipagem.  

Foram realizadas as seguintes etapas de tratamento:

- Conversão de variáveis categóricas codificadas numericamente para rótulos interpretáveis quando necessário para análise visual.
- Padronização de nomes de colunas para garantir consistência e legibilidade.
- Conversão de variáveis numéricas armazenadas como texto para o tipo apropriado.

Essas etapas asseguraram maior confiabilidade e clareza na análise exploratória subsequente.

---

### Análise Exploratória de Dados  

A análise exploratória foi conduzida por meio de histogramas e gráficos de caixa e comparação entre grupos.

#### Variáveis Demográficas  

- **Gênero:** Não foram observadas diferenças significativas no comportamento de evasão entre os gêneros. As proporções de churn se mostraram semelhantes entre as classes.
  
- **Senioridade (idosos acima de 65 anos):** Observou-se uma proporção relativamente maior de churn entre clientes idosos. No entanto, o volume reduzido de clientes nessa faixa etária limita conclusões mais robustas.

- **Presença de parceiro(a):** Clientes sem parceiro(a) apresentam quase o dobro de evasão em termos absolutos, indicando uma tendência relevante.

- **Dependentes:** Foi identificada uma diferença significativa: clientes sem dependentes apresentam um churn aproximadamente cinco vezes maior, mesmo considerando um leve desbalanceamento entre os grupos.

---

#### Serviços Contratados  

- **Serviço telefônico e múltiplas linhas:** Não foram identificados padrões relevantes de evasão.

- **Serviço de internet:** Observou-se uma taxa de churn significativamente elevada entre clientes que utilizam fibra óptica, sugerindo possível insatisfação ou problemas relacionados a esse serviço.

- **Segurança online, backup online e proteção de dispositivos:** Clientes que não possuem esses serviços apresentam maior taxa de evasão. A ausência de segurança online e backup demonstrou associação particularmente perceptível com churn.

- **Assistência técnica:** A ausência desse serviço também está associada a maior evasão.

- **Serviços de streaming (TV e filmes):** Foi identificada uma leve tendência de maior churn entre clientes que não possuem esses serviços, porém o efeito é menos expressivo.

---

#### Tipo de Contrato e Método de Pagamento  

- **Contrato mensal:** Apresenta a maior taxa de churn observada na análise. Clientes com contrato mensal demonstram evasão significativamente superior aos demais tipos de contrato.

- **Faturamento online:** Clientes que optam por faturamento eletrônico apresentam maior evasão.

- **Cheque eletrônico:** Este método de pagamento mostrou-se fortemente associado a altos índices de churn, configurando um dos pontos mais críticos identificados.

---

#### Variáveis Numéricas  

- **Tempo de permanência (tenure):** Clientes com menor tempo de permanência apresentam maior churn, padrão comum em serviços por assinatura.

- **Gasto mensal:** Clientes que evadiram tendem a apresentar gasto mensal mediano mais elevado.

- **Gasto total:** Observou-se valor total menor entre clientes que evadiram, o que está diretamente relacionado ao menor tempo de permanência no serviço.

---

### Conclusões e Insights  

A análise identificou três fatores principais associados à elevada evasão:

1. Contratos mensais  
2. Método de pagamento por cheque eletrônico  
3. Serviço de internet por fibra óptica  

Além disso, a ausência de serviços adicionais (segurança online, backup e proteção) também está associada a maior churn.  

Clientes com menor tempo de permanência representam o grupo mais vulnerável à evasão, indicando a importância de estratégias de retenção nos primeiros meses de contrato.

---

### Recomendações  

Com base nos achados, recomenda-se:

1. **Revisão do modelo de contrato mensal**, incentivando migração para contratos de maior duração por meio de benefícios ou descontos.
2. **Investigação da experiência do cliente no serviço de fibra óptica**, incluindo análise de qualidade, suporte técnico e estabilidade.
3. **Reavaliação do método de pagamento por cheque eletrônico**, considerando simplificação ou incentivos para métodos alternativos.
4. **Criação de pacotes combinados** (internet + segurança online + backup), agregando valor percebido e reduzindo probabilidade de churn.
5. **Estratégias de retenção nos primeiros meses de contrato**, como acompanhamento ativo, ofertas personalizadas ou benefícios progressivos.

A implementação dessas ações pode contribuir significativamente para a redução da taxa de evasão e aumento do valor do ciclo de vida do cliente.
